```python
import pandas as pd
import numpy as np
import random
import string
np.random.seed(56)
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
id()
```

```python
def score():
    return np.random.randint(0, 100)
```

```python
score()
```

```python
df=pd.DataFrame([[score(),score(),score(),score()] for i in range(10)], index=[id() for i in range(1,11)], columns=['국어','영어','수학','사회'])
```

```python
df
```

```python
df.loc[:,'과학'] = pd.Series([score() for i in range(1, 11)], index=df.index)
df
```
